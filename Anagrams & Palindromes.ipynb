```python
## 7 letter anagrams

seven_letter_words = []

for word in word_list:
    if len(word) == 7:
        seven_letter_words.append(word)
        
seven_letter_words[:10]
```

```python
def anagram_checker(value, words):
    anagrams = []
    
    for word in seven_letter_words:
        if sorted(value) == sorted(word):
            anagrams.append(word)
        
    return anagrams

print(anagram_checker('trainer',words))
```
